# K-Means Clustering

## Importing the libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Importing the dataset

```python
data = pd.read_csv('Mall_Customers.csv')
X = data.iloc[:,[3,4]].values
```

## Using the elbow method to find the optimal number of clusters

```python
from sklearn.cluster import KMeans
```

```python
wcs = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcs.append(kmeans.inertia_)
plt.plot(range(1,11),wcs,marker='o')
plt.title('Elbow Method for Optimal KMeans Clusters')
plt.ylabel('WCS score')
plt.xlabel('Number of kmeans clusters')
#plt.show()
```

## Training the K-Means model on the dataset

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42).fit(X)
#creating the dependent variable
y_kmeans = kmeans.fit_predict(X)
```

## Visualising the clusters

```python
plt.scatter(X[:,0],X[:,1])
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1])
```

```python
for i in range(0,5):
    plt.scatter(X[y_kmeans == i,0],X[y_kmeans == i,1],label = 'Cluster %s'%i)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],
           s=200,label='centroids',color = 'pink')
plt.title('Clusters of customers')
plt.ylabel('Spending score (1-100)')
plt.xlabel('Annual income (k$)')
plt.legend()
```
